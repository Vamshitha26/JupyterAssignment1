# IMBD EDA PROJECT

```python
#Libraries we need for the project
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("imdb_movies.csv") #reading the file
df
```

```python
df.info() # checking if there are any null values
```

```python
df.isnull().sum() #count of null values - genre (85), crew(56) -These are to replaced with unavailable 
```

```python
df["date_x"]= pd.to_datetime(df["date_x"]) # To change the datatype of the date column to datetime data type.
```

```python
df.info()
```

```python
# Filling null values with "unavailable"
df["genre"] = df["genre"].fillna("Unavailable")
df["crew"] = df["crew"].fillna("Unavailable")
```

```python
df.info() #All the null values have been removed
```

```python
df["year"] = df["date_x"].dt.strftime("%Y")
df
```

```python
sns.histplot(df["year"])
plt.show()
```

```python
gb = df.groupby("genre").agg({"date_x":"count"})
gb = gb.sort_values (by = "date_x", ascending = False)
gb = gb.head(10)

```

```python
sns.barplot(x=gb.index, y = gb["date_x"], data = gb)
plt.xticks(rotation = 45)
plt.show()
```

```python
#Is there a relationship between a movie’s runtime and its rating? Plot a scatter plot and describe any observed trend.
plt.figure(figsize = (12,4))
sns.scatterplot(x="year", y="score", data = df)
plt.xticks(rotation = 45, fontsize = 6)
plt.show()
#The number of ratings have increased all in 2000's than on previous years. Also in 1900 there are no movies between 0-30 ratings.
```

```python
#how do ratings vary by genre? Use a boxplot to visualize the differences in ratings across genres.
sns.boxplot(gb.index)
plt.show()
```

```python
#Is there a correlation between the number of votes a budget and revenue? Create a scatter plot and calculate the correlation coefficient. What can you conclude?
plt.figure(figsize = (12,4))
sns.scatterplot(x="budget_x", y="revenue", data = df)
plt.xticks(rotation = 45, fontsize = 6)
plt.show()
```

```python
#Which genre has the highest average rating? Calculate the average rating for each genre and plot the results.
gb = df.groupby("genre").agg({"score":"mean"})
plt.figure(figsize = (12,12))
sns.barplot(x="score",y="genre", data=gb)
plt.xticks(rotation = 45)
plt.gca().invert_yaxis()
plt.show()


```

```python
#How does the popularity of genres vary over time? Plot the number of movies released per genre each year.
movrel = df.groupby(["genre", "year"]).agg({"status":"count"})
plt.figure(figsize = (20,20))
sns.barplot(x="year",y="genre", data=df, hue ="status")
plt.gca().invert_yaxis()
plt.show()
```

```python
#How has the average movie rating changed over the years? Plot the average rating for each year.
ar = df.groupby("year").agg({"score":"mean"})
plt.figure(figsize = (20,10))
sns.barplot(x="year",y="score",data = ar)
plt.xticks(rotation = 45, fontsize = 6)
plt.show()
```

```python
#Which years had the highest and lowest number of movie releases? Plot the number of movies released each year.
movrelyr = df.groupby(['year', 'status']).agg({'date_x': 'count'}).reset_index()
plt.figure(figsize = (25,10))
sns.barplot(x="year",y="date_x",data=movrelyr,hue="status")
plt.show()




```

```python
#Which genres are most popular in each decade? Create a bar plot showing the most frequent genres by decade.
df["yearint"]= df.year #ccreating new column
df["yearint"] = pd.to_numeric(df["yearint"]) #converting year column to int
df['decade'] = (df['yearint'] // 10) * 10 #converting years into decades
gendec = df.groupby(['decade', 'genre']).agg({'date_x': 'count'})
plt.figure(figsize = (25,10))
plt.scatter(df["genre"],df["date_x"],c=df["decade"], cmap ="summer")
plt.colorbar()
plt.xticks(rotation = 45, fontsize = 6)
plt.show()

```

```python
#Plot a heatmap or pairplot to examine relationships between budget, revenue, scores.
corr = df[['budget_x', 'revenue']].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.show()
```

```python
#Are there specific genres or release years with higher-rated movies? Group by genre and year, then analyze the average rating.
relyear = df.groupby(["genre","year"]).agg({"score":"mean"})
ascen = relyear.sort_values(by="score", ascending = False)
ascen.head(10) #Top 10 with highest scores with genre and years mentioned.
```
